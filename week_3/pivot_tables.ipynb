# Pivot Tables

A pivot table is a way of summarizing data from a dataframe. It makes heavy use of the aggregation function. A pivot table itself is a dataframe where the cells are some aggregate value based on another dataframe.

It also tends to include some marginal values such as the sums for each column and row. This allows you to see the relationship between two variables with just a glance.

Load a new dataset on cars.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/cars.csv')
df.head()
```

Using a pivot table, pivot out one of the columns into column headers, and compare it to another column as row indices. For instance, compare the makes of electric vehicles vs. the years. Do this comparison in terms of battery capacity.

In plainer english, what is happening here is that the battery capacity in kW is being averaged for each combination of make and year.

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc = np.mean)
```

What can we get from this table? A few things:

* Some makes did not start producing electric cars until later years. Kia's first model came in 2015, while Mitsubishi and Nissan had models in 2012.
* The battery capacity increased dramatically for Tesla, but not for anyone else
* The battery capacities vary greatly across the makes in 2016.

I can pass `pandas` several functions to aggregate. I can also ask `pandas` to give me a summary, or `All` row and column.

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc = [np.mean,np.min], margins = True)
```
