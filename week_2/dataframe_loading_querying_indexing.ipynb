# DataFrame Loading, Querying, Indexing

A common workflow:

* Read data into a dataframe, usually from an external file

```python
import pandas as pd
import os
```

# Loading data

Pandas has built-in support for delimited files such as `.csv` files. We will look at data about olympic medals in the `../data` folder.

```python
df = pd.read_csv('../data/olympics.csv')
df.head()
```

It looks like the first row of data should be the column names, and the first column with country names should be the indices. Use input parameters for `read_csv` to make this happen.

```python
df = pd.read_csv('../data/olympics.csv', index_col = 0, skiprows = 1)
df.head()
```

This data was imported from `https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table`, and the column labels for gold, silver, and medal came through as `01 !`, etc. We want to modify those using `.rename`.

```python
for col in df.columns:
    print(col, col[4:])
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2] == '02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2] == '03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
```

```python
for col in df.columns:
    print(col)
```

```python
df.head()
```

# Querying a DataFrame

A Boolean mask is an array where each value is either True or False. It is overlaid on top of a data structure that we're querying, and any cell aligned with a True value will be retained in our final result.

Boolean masks are created by applying operators directly to a pandas DataFrame. 

Create a Boolean mask that shows which countries have received at least one summer gold medal.

```python
df['Gold'] > 0
```

Now overlay that mask on the DataFrame using the `where` function.

```python
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

Here, all of the countries that did not meet the condition have `NaN` data instead. 

Count how many entries there are (how many countries have a summer gold).

```python
only_gold['Gold'].count()
```

How many countries were there originally?

```python
df['Gold'].count()
```

How can we drop the rows that don't have any data? Use the `drop_na` function. 

```python
only_gold = only_gold.dropna()
only_gold.head()
```

This is a more direct way of querying the DataFrame:

```python
df[df['Gold']>0].head()
```

Now be more generous. Look for countries that have received a gold in the summer OR winter games (working from the summer and winter columns, not from the total column). 

Note that each Boolean mask has to be encased in parenthesis.

```python
df[(df['Gold'] > 0) | (df['Gold.1'] > 0)].head()
```

```python
len(df[(df['Gold'] > 0) | (df['Gold.1'] > 0)])
```

So there is one country who has only won a gold in the winter olympics, never the summer olympics. Which country is it? 

```python
df[(df['Gold'] == 0) & (df['Gold.1'] > 0)]
```

# Indexing DataFrames

## Basics

```python
df.head()
```

What if we don't want to index by the name of the country, but rather by the number of gold medals that were won at summer games?

Generate a new column to preserve the country info.

```python
df['country'] = df.index
df = df.set_index('Gold')
df.head()
```

It looks like there is an empty row, but actually the index has a name. Also notice that the `Gold` column is gone. 

We can get rid of the index entirely by using the function `reset_index()`, which changes the indices to integers and moves the current index values to a column. 

```python
df = df.reset_index()
df.head()
```

## Multi-level indexing

Pandas has the option to do multi-level indexing. We simply call set index and give a list of columns. This is particularly useful for geographical data which has many layers of geographic regions.

Look at some census data.

```python
df = pd.read_csv('../data/census.csv')
df.head()
```

```python
len(df)
```

There is a lot of data here. 3193 rows and 100 columns!

```python
df.columns
```

This contains summaried data on three levels: The whole country, each state, and each county. 

What if we want to see all of the unique values in each column? 

```python
df['SUMLEV'].unique()
```

Reduce the DataFrame a bit and only keep some of the data. Only look at data where `SUMLEV == 50`. These are the summaries at the state level (remove those at the county level).

```python
df = df[df['SUMLEV']==50]
df.head()
```

Reduce the data to the total number of population estimates and the total number of births. Specify which columns we want to keep. 

```python
columns_to_keep = ['STNAME',
                  'CTYNAME',
                  'BIRTHS2010',
                  'BIRTHS2011',
                  'BIRTHS2012',
                  'BIRTHS2013',
                  'BIRTHS2014',
                  'BIRTHS2015',
                  'POPESTIMATE2010',
                  'POPESTIMATE2011',
                  'POPESTIMATE2012',
                  'POPESTIMATE2013',
                  'POPESTIMATE2014',
                  'POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

Set the index of the data to the state and county names. This produces a dual index- first the state name, and then the county name. 

```python
df = df.set_index(['STNAME','CTYNAME'])
df.head()
```

How do we query this DataFrame? We must provide the arguments in order by the order in which we wish to query.

Look at data from one county (our home county!):

```python
df.loc['Michigan','Washtenaw County']
```

Look at data across two counties. Provide both pairs of indices. 

```python
df.loc[[('Michigan','Washtenaw County'),('Michigan','Wayne County')]]
```

This is an example of hierarchical labels by rows. You can also do hierarchical labels by columns. 

## Challenge question

Reindex the purchase records DataFrame to be indexed hierarchically, first by store, then by person. Name these indexes 'Location' and 'Name'. Then add a new entry to it with the value of:

Name: 'Kevyn', Item Purchased: 'Kitty Food', Cost: 3.00, Location: 'Store 2'.

```python
purchase_1 = pd.Series({'Name': 'Chris',
                        'Item Purchased': 'Dog Food',
                        'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn',
                        'Item Purchased': 'Kitty Litter',
                        'Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod',
                        'Item Purchased': 'Bird Seed',
                        'Cost': 5.00})

df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])

df.head()
```

First turn the current indices- the store names- into a column.

```python
df = df.reset_index()
df = df.rename(columns = {'index':'Location'})
df.head()
```

```python
df = df.set_index(['Location','Name'])
df.head()
```

```python
df = df.append(pd.Series({'Cost':3.00,'Item Purchased':'Kitty Food'}, name=('Store 2','Kevyn')))
df
```

# Handling Missing Values

In other fields, it is common to fill missing values with `99`. Pandas has the ability to fill it with `NaN`. 

Look at the file `../data/log.csv`.

```python
df = pd.read_csv('../data/log.csv')
df
```

We can use the pandas function `fillna` to manage the `NaN` values. Use the method `ffill` for forward filling (fill data based on a previous row). 

```python
df = df.set_index('time')
df = df.sort_index()
df
```

Indices are not unique, and several users can use the system at the same time. To manage this, use multi-level indexing of time and user name. 

```python
df = df.reset_index()
df = df.set_index(['time','user'])
df
```

A lot of the pandas packages are written to handle missing values. For instance, if you use `numpy.mean`, it will exclude any missing values.
